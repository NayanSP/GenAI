```python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_classic.vectorstores import Pinecone
from langchain_classic.text_splitter import RecursiveCharacterTextSplitter
from langchain_classic.document_loaders import PyPDFDirectoryLoader
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
def read_doc(directory):
    file_loader = PyPDFDirectoryLoader(directory)
    documents = file_loader.load()
    return documents
```

```python
doc = read_doc('budget_speech.pdf')
len(doc)
```

```python
def ChunkData(docs, chunk_size=1000, chunk_overlap=100):
    text_split = RecursiveCharacterTextSplitter(chunk_size, chunk_overlap)
    doc = text_split.split_documents(docs)
    return doc
```

```python
document = ChunkData(docs=doc)
len(document)
```
